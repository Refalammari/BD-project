## 3 - Preproccing

### 3.1 - Dropping not needed columns

```python
df = df.drop(columns=[df.columns[0], 'salary', 'salary_currency'])
```

### 3.2 - Checking for nulls



```python
import matplotlib.pyplot as plt # Added import statement for matplotlib.pyplot
import seaborn as sns
fig, ax = plt.subplots(figsize=(9,5))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.show()
```

### 3.3 - Encoding nominal attributes


```python
from sklearn.preprocessing import LabelEncoder
# Columns to encode
columns_to_encode = ['experience_level', 'employment_type', 'job_title', 'employee_residence', 'company_location', 'company_size']

# Initialize LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to each of the specified columns
for column in columns_to_encode:
    df[column] = le.fit_transform(df[column])

# Check the transformed dataset
print(df.head())


```

### 3.4 - Normalizing salary

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize 'salary_in_usd' using MinMaxScaler
df['salary_in_usd'] = scaler.fit_transform(df[['salary_in_usd']])

# Check the transformed salary column
print(df[['salary_in_usd']].head())
```

### 3.5 - Exploring colleration for salary with the rest of the attributes

```python
import seaborn as sns
# Calculate the correlation matrix
NumFeatures = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = NumFeatures.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
